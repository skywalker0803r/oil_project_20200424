```python
import pandas as pd
import os
import numpy as np
from tqdm import tqdm_notebook as tqdm
import joblib
pd.set_option('display.max_columns', None)
root = '../data/phase_2/raw/other/'
os.listdir(root)
```

```python
y_cols = ['C5NP','C5IP','C5N',
           'C6NP','C6IP','C6N','C6A',
           'C7NP','C7IP','C7N','C7A',
           'C8NP','C8IP','C8N','C8A',
           'C9NP','C9IP','C9N','C9A',
           'C10NP','C10IP','C10N','C10A']
```

# load dataset

```python
df = pd.read_excel(root+'FCFC Import Naphtha Composition Lab_001~100_R2.xlsx',sheet_name='001~100')
df = df.dropna(axis=0,how='all')
df = df.dropna(axis=1,how='all')
df = df.reset_index(drop=True)
df.columns = [i for i in range(df.shape[1])]
```

# y23

```python
y23 = df.loc[33:33+22,1:].reset_index(drop=True).set_index(1).T.reset_index(drop=True).head(23).astype('float')
y23.columns = y_cols
```

```python
y23.head(3)
```

# y54

```python
y54 = df.loc[64:64+54-1,1:].set_index(1).T.reset_index(drop=True).head(23).astype('float')
```

```python
y54.head(3)
```

# W = inv(X)@Y

```python
W = np.linalg.inv(y23.values)@y54.values
print(W.shape)
```

# build transform23to54

```python
class transform_2354(object):
    def __init__(self):
        self.x_cols = y23.columns.tolist()
        self.y_cols = y54.columns.tolist()
        self.W = W
    
    def __call__(self,x):
        res = x.values@self.W
        return pd.DataFrame(res,columns=self.y_cols)
```

```python
tr = transform_2354()
```

# make pred54

```python
y54_pred = tr(y23)
```

# check ouput

# save TR

```python
joblib.dump(tr,'../model/transformer(23_to_54).pkl')
print('save done!')
```
