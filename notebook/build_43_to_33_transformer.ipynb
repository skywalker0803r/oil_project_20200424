# modeling

```python
from tqdm import tqdm_notebook as tqdm
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression

class model4333(object):
    def __init__(self,x_col,y_col):
        self.x_col = x_col
        self.y_col = y_col
        self.model = {}
        for y_name in self.y_col:
            self.model[y_name] = LinearRegression()
            
    def fit(self,X,y):
        for name in tqdm(self.y_col):
            self.model[name].fit(X,y[name])
        print('fit done')
    
    def predict(self,X):
        result = pd.DataFrame(index=X.index,columns=self.y_col)
        for name in self.y_col:
            result[name] = self.model[name].predict(X)
        return result
```

```python
m = model4333(x_col,y_col)
```

```python
import warnings;warnings.simplefilter('ignore')
m.fit(X_train,y_train)
```

```python
# make prdiction
y_pred = y_test.copy()
y_pred[:] = m.predict(X_test)
```

# metrics

```python
def mape(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
res = pd.DataFrame(index=y_col,columns=['MAPE'])
for i in y_col:
    res.loc[i,'MAPE'] = mape(y_pred[i],y_test[i])
res = res.reset_index()
res
```
