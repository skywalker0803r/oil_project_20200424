```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
simulation['進料量'] = train[col_names['fna']].values.ravel()
real['進料量'] = pd.to_numeric(dcs['進料量'],errors='coerce').dropna().values

simulation['回流量'] = train['Yield Summary_Reflux Rate_m3/hr'].values.ravel()
real['回流量'] = pd.to_numeric(dcs['回流量'],errors='coerce').dropna().values

simulation['輕質流量'] = train['Light End Product Properties_Light End Prod. Rate_m3/hr'].values.ravel()
real['輕質流量'] = pd.to_numeric(dcs['LE m3/hr'],errors='coerce').dropna().values

simulation['重質流量'] = train['Heavy End Product Properties_Heavy End Product Rate_m3/hr'].values.ravel()
real['重質流量'] = pd.to_numeric(dcs['HE m3/hr'],errors='coerce').dropna().values

simulation['重質流量'] = train['Heavy End Product Properties_Heavy End Product Rate_m3/hr'].values.ravel()
real['重質流量'] = pd.to_numeric(dcs['HE m3/hr'],errors='coerce').dropna().values

simulation['塔底溫度'] = train['Column Temp Profile_C013 Tray 16 Temp_oC'].values.ravel()
real['塔底溫度'] = pd.to_numeric(dcs['Unnamed: 28'],errors='coerce').dropna().values

simulation['塔頂溫度'] = train['Column Temp Profile_C011 Tray 15 Temp_oC'].values.ravel()
real['塔頂溫度'] = pd.to_numeric(dcs['Unnamed: 22'],errors='coerce').dropna().values

#simulation['加熱爐'] = train['Duty_Reboiler Heat Duty_Mkcal/hr'].values.ravel()
#real['加熱爐'] = pd.to_numeric(dcs['Unnamed: 22'],errors='coerce').dropna().values

#simulation['冷凝器'] = train['Column Temp Profile_C011 Tray 15 Temp_oC'].values.ravel()
#real['冷凝器'] = pd.to_numeric(dcs['Unnamed: 22'],errors='coerce').dropna().values
```

```python
for i in real.keys():
    print(i)
    sns.kdeplot(simulation[i],label='simulation')
    sns.kdeplot(real[i],label='real')
    plt.show()
```
