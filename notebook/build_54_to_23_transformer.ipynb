```python
import pandas as pd
import os
import numpy as np
from tqdm import tqdm_notebook as tqdm
import joblib
pd.set_option('display.max_columns', None)
root = '../data/phase_2/raw/other/'
os.listdir(root)
```

```python
y_cols = ['C5NP','C5IP','C5N',
           'C6NP','C6IP','C6N','C6A',
           'C7NP','C7IP','C7N','C7A',
           'C8NP','C8IP','C8N','C8A',
           'C9NP','C9IP','C9N','C9A',
           'C10NP','C10IP','C10N','C10A']
```

# load dataset

```python
df = pd.read_excel(root+'FCFC Import Naphtha Composition Lab_001~100_R2.xlsx',sheet_name='001~100')
df = df.dropna(axis=0,how='all')
df = df.dropna(axis=1,how='all')
df = df.reset_index(drop=True)
df.columns = [i for i in range(df.shape[1])]
```

# y23

```python
y23 = df.loc[33:33+22,1:].reset_index(drop=True).set_index(1).T.reset_index(drop=True).head(23).astype('float')
y23.columns = y_cols
y23.head(3)
```

# y54

```python
y54 = df.loc[64:64+54-1,1:].set_index(1).T.reset_index(drop=True).head(23).astype('float')
y54.head(3)
```

# X@W = Y

# W = inv(X)@Y

```python
W = np.linalg.inv(y23.values)@y54.values
print(W.shape)
```

```python
W = pd.DataFrame(W,index=y23.columns,columns=y54.columns)
W = W.apply(lambda x:round(x,3))
W.style.background_gradient(axis=0)
```

# how? 54 to 23

```python
a = y23
b = (y54/W.sum(axis=0).values).fillna(0)@W.T.values
b.columns = y23.columns
```

```python
a.head()
```

```python
b.head()
```

# check is correct

```python
for i in y23.columns:
    if np.allclose(a[i].values,b[i].values) == False:
        w = W.loc[i].values
        print(i,w[w>0.00001])
```

# build transform54to23

```python
class transform_5423(object):
    def __init__(self):
        self.x_cols = y54.columns.tolist()
        self.y_cols = y23.columns.tolist()
        self.W = W
    
    def __call__(self,x):
        y23 = (x/self.W.sum(axis=0).values).fillna(0)@self.W.T.values
        y23.columns = self.y_cols
        return y23
```

```python
tr = transform_5423()
```

# make pred54

```python
y23_pred = tr(y54)
```
