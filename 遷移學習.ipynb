```python
import os
from API.utils import *
import joblib
os.listdir('./data/phase_2/cleaned')
```

```python
real_dict = joblib.load('./data/phase_2/cleaned/real_dict.pkl')
```

```python
col_names = joblib.load('./data/phase_2/cleaned/col_names.pkl')
```

```python
m58_162 = joblib.load('./model/ANN(58_to_sp162).pkl')#分離係數模組
m162_162 = joblib.load('./model/transformer(SP162_to_Y162).pkl')#從分離係數還原組成
m54_33 = joblib.load('./model/transformer(54_to_33).pkl')#54轉33
ANN_C = joblib.load('./model/ANN(tray).pkl')#神經網路(C)負責預測熱量以及流量
ANN_B = joblib.load('./model/ANN(energy).pkl')#神經網路(B)負責預測密度
```

```python
real_dict.keys()
```

```python
real_df = pd.DataFrame(columns=m58_162.x_col)
real_df
```

```python
real_df['Case Conditions_Feed Rate to C013 (Input)_m3/hr'] = real_dict['旁通流量']
real_df['Case Conditions_C7+ in Light End (Input)_wt%'] = real_dict['C7+ in Light End (Input)']
real_df['Case Conditions_C6P- in Heart Cut (Input)_wt%'] = real_dict['C6P- in Heart Cut (Input)']
real_df['Case Conditions_Heart Cut Prod. Rate (Input)_m3/hr'] = real_dict['核心流量']
real_df.iloc[:,4:] = real_dict['xna'].values
real_df
```

```python
def 操作建議模組(x):
    sp162 = m58_162.predict(x)
    xna = x[col_names['xna']].astype('float32')
    xhc = m162_162(xna,sp162)[col_names['xhc']]
    return m54_33(xhc),ANN_B.predict(x),ANN_C.predict(x)
    
```

```python
xhc,b,c = 操作建議模組(real_df)
```

```python
xhc
```

```python
real_dict['重組進料33組成']
```

```python
from sklearn.metrics import mean_squared_error

res = pd.DataFrame(index=xhc.columns,columns=['MSE'])
for i in xhc.columns:
    res.loc[i,'MSE'] = mean_squared_error(xhc[i],real_dict['重組進料33組成'][i])
res.loc['AVG'] = res.mean(axis=0)
res
```

```python
x_train = xhc.iloc[:100,:]
x_test = xhc.iloc[100:,:]
y_train = real_dict['重組進料33組成'].iloc[:100,:]
y_test = real_dict['重組進料33組成'].iloc[100:,:]
```

```python
from xgboost import XGBRegressor
from sklearn.multioutput import MultiOutputRegressor
xgb = MultiOutputRegressor(XGBRegressor())
xgb
```

```python
xgb.fit(x_train,y_train)
```

```python
y_pred = xgb.predict(x_test)
```

```python
y_pred = pd.DataFrame(y_pred,columns=y_test.columns)
```

```python
res = pd.DataFrame(index=y_pred.columns,columns=['MSE'])
for i in y_pred.columns:
    res.loc[i,'MSE'] = mean_squared_error(y_pred[i],y_test[i])
res.loc['AVG'] = res.mean(axis=0)
res
```
