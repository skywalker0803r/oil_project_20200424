# 模型預測ｗｔ，分離係數

```python
df_wt_predict,df_sp_predict = eva.C.predict(df_X)
```

```python
df_sp_predict.index = df_X.index
df_wt_predict.index = df_X.index
```

```python
c6_col = eva.C.net.c6_col
c7_col = eva.C.net.c7_col
df_wt = df[df_wt_predict.columns]
```

```python
table = pd.DataFrame()
table['預測c6p'] = df_wt_predict.iloc[:,c6_col].sum(axis=1).values
table['真實c6p'] = df_wt.iloc[:,c6_col].sum(axis=1).values
table['預測c7p'] = df_wt_predict.iloc[:,c7_col].sum(axis=1).values
table['真實c7p'] = df_wt.iloc[:,c7_col].sum(axis=1).values
table
```

```python
from sklearn.metrics import mean_squared_error
print(mean_squared_error(table['預測c6p'],table['真實c6p']))
print(mean_squared_error(table['預測c7p'],table['真實c7p']))
```

# 模型預測密度

```python
E_pred = eva.E.predict(df_X)
```

```python
E_pred.head()
```

```python
density = E_pred[['NA_density','LE_density','HC_density','HE_density']]
```

```python
shc = df_sp_predict[col_names['shc']].values
sle = df_sp_predict[col_names['sle']].values
she = df_sp_predict[col_names['she']].values
xna = df[col_names['xna']].values
```

```python
fhc_ton = 100 * density['HC_density'].values
fle_ton = fhc_ton * (xna*sle).sum(axis=1) / (xna*shc).sum(axis=1)
fhe_ton = fhc_ton * (xna*she).sum(axis=1) / (xna*shc).sum(axis=1)
```

```python
table = pd.DataFrame()
table['真實fle'] = df[col_names['fle']].values.ravel()
table['預測fle'] = fle_ton
table['真實fhc'] = df[col_names['fhc']].values.ravel()
table['預測fhc'] = fhc_ton
table['真實fhe'] = df[col_names['fhe']].values.ravel()
table['預測fhe'] = fhe_ton
table
```

```python
for i in ['fle','fhc','fhe']:
    print(i,'mse',mean_squared_error(table['真實{}'.format(i)],table['預測{}'.format(i)]))
```
